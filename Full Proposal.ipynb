# Abstract

## Overview and Motivation

This research investigates whether there is an association between the number of rest days between games and the likelihood of injuries among NBA players across different teams over the yearly seasons of the league. Injuries are a pivotal factor in determining team performance, especially during the high-stakes NBA playoffs. Previous studies have highlighted how workload management and rest days influence injury risk, but there is a gap in quantifying this relationship across entire seasons. By examining this association, we aim to provide quantifiable and data-proven insights for optimizing player recovery and reducing injury risks, ultimately improving team performance and the well being of players for the future.

## Methodology

The study uses two datasets: (1) NBA Injury Stats (1951–2023), which details injuries by date, team, and player status, and (2) NBA Rest Days (2012–2024), which outlines the frequency of rest days for each team in the league. After preprocessing and cleaning the data, including aligning years, handling missing values, and consolidating team names across seasons, the datasets were merged to create a comprehensive database. Statistical analysis was conducted to identify correlations between rest days, injury rates, and team performance metrics (e.g., win percentages). Regression models and hypothesis testing were employed to validate associations and quantify the impact of rest days on injury prevalence and game outcomes.

## Findings

The analysis identifies key trends in injury reporting over the years. During the early decades (1950s–1980s), the number of injuries remained low, reflecting smaller league sizes, fewer games, and limited record-keeping practices. From the 1990s onward, a sharp rise in injuries aligns with league expansion, increased game intensity, and improved documentation. Recent fluctuations (2020–2021) coincide with the COVID-19 pandemic, which disrupted game schedules and team dynamics. Regression analysis confirms a long-term upward trend in injuries across all teams, emphasizing the need to explore contributing factors, such as player workload and health management practices. Team-level analysis highlights injury disparities, with the San Antonio Spurs, Boston Celtics, and Dallas Mavericks recording the highest counts. Further, games played under high-intensity schedules (e.g., back-to-back games and four games in five days) negatively impact performance, as reflected in lower win percentages and average efficiency differentials (AED). Correlation analysis reveals that longer rest periods significantly improve team performance, underscoring the importance of optimizing schedules to balance competitiveness and player welfare. Ethical considerations acknowledge the biases inherent in injury reporting and the limitations of publicly available data. Despite these constraints, this study contributes to understanding the relationship between injuries, performance, and league-wide scheduling policies, offering actionable insights for injury prevention and improved player health.

## Conclusions

This study highlights the critical role of rest days in reducing injuries and enhancing team performance, measured by win rate (percentage), in the NBA. The findings suggest that scheduling adjustments, such as increasing the number of rest days between games, could be a practical strategy to mitigate injury risks. Future work could explore other factors, such as travel schedules and player workload, to develop a holistic understanding of injury prevention in professional basketball. These insights could also have broader implications and improvements for athlete management in other high-intensity sports besides basketball.

## Dataset #1 (NBA Injury Stats 1951 - 2023)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
injuries = pd.read_csv('NBA Player Injury Stats(1951 - 2023).csv')
injuries['Date'] = pd.to_datetime(injuries['Date'])
```

```python
team_name_mapping = {
    'Bullets': 'Wizards',
    'Zephyrs': 'Wizards',
    'Cavailers': 'Cavaliers',
    'Braves': 'Clippers',  
    'Sonics': 'Thunder',       
    'Colonels (ABA)': 'Colonels',
    'Squires (ABA)': 'Squires',
    'Chaparrals (ABA)': 'Chaparrals',
    'Tams (ABA)': 'Tams',
    'Nets (ABA)': 'Nets',
    'Spirits (ABA)': 'Spirits',
    'Stars (ABA)': 'Stars',
    'Nuggets (ABA)': 'Nuggets'
}
injuries['Team'] = injuries['Team'].replace(team_name_mapping)
injury_events = injuries[(injuries['Acquired'].notna()) | (injuries['Relinquished'].notna())].copy()
injury_events.sort_values(by=['Team', 'Date'], inplace=True)
injury_events['Rest Days'] = injury_events.groupby('Team')['Date'].diff().dt.days
injury_events['Rest Days'] = injury_events['Rest Days'].fillna(0)
injury_events = injury_events.drop(columns=['Unnamed: 0'])
injury_events
```

## Line Plot: Number of Injuries Per Year

```python
# Count number of injuries per year
injuries_per_year = injuries.groupby('Year').size().reset_index(name='Injuries')

plt.figure(figsize=(12, 6))
injuries_per_year.plot(kind='line', x='Year', y='Injuries', title='Number of Injuries Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Injuries')
plt.grid()
plt.show()
```

## **Results and Interpretation**

The resulting line plot reveals the following key insights:

### **Early Years (1950s-1980s)**
The number of injuries was relatively low during this period. This may be attributed to a few things, including:
- Fewer games played during this period, due to smaller league sizes with fewer teams and players. The NBA only had 10 or fewer teams all the way until 1967, and never had more than 22 teams until 1976.
- Limited record-keeping practices compared to modern standards.

### **Recent Decades (1990s-2020s)**
We see an overall large increase from 1990 to 2020, as well as three sharp increases here. The first being around 1992-1993, the second being around 2002-2003, and the last one being around 2015-2016. A few factors that contributed to these increases include: 
- An expanded league with more teams and games per season. The NBA expanded to 22 teams in 1976 (previously at 18), and continued to gradually expand until reaching 30 teams in 2004. 
- Improved injury reporting and documentation systems.
- Increased physical demands on players due to game intensity and competition.

### **Recent Fluctuations (2020s)**
The dip around 2020-2021 aligns with the COVID-19 pandemic, which reduced the number of games as the country went into a state of lockdown.



The trend of injuries shows a significant rise over time, emphasizing the need for further exploration of contributing factors, such as player workload, rest schedules, and injury prevention strategies. This analysis provides a foundation for evaluating the impact of injuries on team performance and player careers.
